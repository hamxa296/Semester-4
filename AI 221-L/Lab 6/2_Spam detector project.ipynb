# Using CountVectorizer class to split each word and treat them separately

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer( max_features=500,   # only top 500 words
    min_df=10)
```

```python
y=df["label"]
y
```

```python
y.value_counts().unique()
```

```python
x=df["sms"]
x
```

```python
x=cv.fit_transform(x)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
    "AI lab teaches deep learning",
    "Deep learning improves AI systems",
    "AI systems use models"
]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
print(X)  # sparse representation
print(X.toarray())  # full dense matrix
print(vectorizer.get_feature_names_out())
```

# Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Train Model using NaiveBayes

```python
from sklearn.naive_bayes import MultinomialNB,BernoulliNB, GaussianNB
```

```python
model=MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix ,classification_report
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
