**Objective:** Detect fraudulent transactions in a highly imbalanced dataset.

**Models to implement:** Logistic Regression, Support Vector Machine (SVM), Artificial Neural Network (ANN / MLPClassifier).

*Run the cell below to generate the imbalanced dataset (95% normal, 5% fraud).*

```
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification

# Generate imbalanced data
X_data, y_data = make_classification(n_samples=5000, n_features=10, n_informative=5, 
                                     n_redundant=2, weights=[0.95, 0.05], random_state=42)

df = pd.DataFrame(X_data, columns=[f'Feature_{i}' for i in range(1, 11)])
df['Class'] = y_data

df['Class'].value_counts()
```
