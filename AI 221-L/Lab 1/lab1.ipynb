```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = {
    "Hours": [1,2,3,4,5],
    "Marks": [35,40,50,55,65]
}

df = pd.DataFrame(data)
print(df)
```

```python
x = df["Hours"].values
y = df["Marks"].values

print("X:", x)
print("Y:",y)
```

```python
plt.scatter(x,y)
plt.xlabel("Hours studied")
plt.ylabel("Marks Obtained")
plt.title("Hours vs Marks")
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

print("Mean of x:", x_mean)
print("Mean of y:",y_mean)
```

```python
numerator = np.sum((x-x_mean)*(y-y_mean))
denominator = np.sum((x-x_mean)**2)

a = numerator/denominator
print("Slope (a):",a)
```

```python
b = y_mean-a*x_mean
print("Intercept (b):", b)
```

```python
print(f"Regression Equation: y = {a:.2f}x + {b:.2f}")
```

```python
y_pred = a*x+b
print("Predicted Values:" , y_pred)
```

```python
z = np.array([4,5,6,7])

y_pred_z = a*z+b

print("Predicted Values for z:", y_pred_z)
```

```python
plt.scatter(x,y,label="Actual Data")
plt.plot(x,y_pred,color="red",label = "Regression Line")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.legend()
plt.show()
```

```python
errors = y - y_pred
print("Residuals:", errors)
```

```python
mse = np.mean(errors **2)
print("Mean Squared Error:",mse)
```
