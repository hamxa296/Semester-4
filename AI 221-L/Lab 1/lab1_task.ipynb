```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("advertising_vs_sales.csv")
df.head()
```

```python
x = df["Ad_Spend_Thousand_PKR"].values
y = df["Monthly_Sales_Thousand_PKR"].values
```

```python
print("X:", x)
print("Y:",y)
```

```python
plt.scatter(x,y)
plt.xlabel("Expenditure on Ads (1000 PKR)")
plt.ylabel("Monthly Sales (1000 PKR)")
plt.title("Monthly Sales vs Ad Expenditure")
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

print(f"Mean of x: {x_mean:.2f}")
print(f"Mean of y: {y_mean:.2f}")
```

```python
num = np.sum((x-x_mean)*(y-y_mean))
denom = np.sum((x-x_mean)**2)

m = num/denom
print(f"Slope (m):{m:.2f}")
```

```python
b = y_mean-m*x_mean
print(f"Intercept (b): {b:.2f}")
```

```python
print(f"Regression Equation: y = {m:.2f}x + {b:.2f}")
```

```python
y_pred = m*x+b
d_y_pred = np.round(y_pred)
print("Predicted Values:" , d_y_pred)
```

```python
plt.scatter(x,y, label="Original Data", color="#FF0000", marker="*")
plt.xlabel("Expenditure on Ads (1000 PKR)")
plt.ylabel("Monthly Sales (1000 PKR)")
plt.plot(x,y_pred,label="Predicted",color="#0CDAE5A9")
plt.legend()
plt.show()
```

```python
errors = y - y_pred
d_errors=np.round(errors)
print("Residuals:", d_errors)
```

```python
mse = np.mean(errors **2)
print(f"Mean Squared Error: {mse:.2f}")
```
